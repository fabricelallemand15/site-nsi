````pycon
>>>

. . .

````

## Corrig exercice 3

`````python
somme = 500
annee = 2017
while somme < 1000:
    somme = somme * (1+3.2/100)
    annee = annee + 1
    print(f"En {annee}, le capital sera de {somme} .")
`````

En complment, on peut demander que l'affichage se fasse avec deux dcimales (cela ne change pas la valeur de la variable qui conserve, elle, toutes ses dcimales).

`````python
somme = 500
annee = 2017
while somme < 1000:
    somme = somme * (1+3.2/100)
    annee = annee + 1
    print(f"En {annee}, le capital sera de {somme:0.2f} .")
`````

## Exercice 4 (nonc et corrig)

On donne ci dessous quatre programmes comportant une boucle `while`. Ces boucles se terminent-elles ou sont-elles des boucles infinies ?

````py
# Programme 1
nombre = 10
while nombre!=27 :
    nombre = nombre+2
````

Le programme 1 ne se termine pas car la variable `nombre` ne prendra que des valeurs paires et ne sera donc jamais gale 27. La boucle est infinie.

````py
# Programme 2
i = 0
while i<10 :
    print(i+1)
````

Le programme 2 ne se termine pas car la valeur de la variable `i` n'est pas modifie dans la boucle : elle reste toujours gale  0, elle ne deviendra donc jamais suprieure  10. La boucle est infinie.

````py
# Programme 3
i = 2
while i<10 :
    i = 0
    i = i+1
````

Le programme 3 ne se termine pas car la valeur de la variable `i` est remise  0  chaque passage dans la boucle et prend donc ensuite la valeur 1. Elle ne deviendra donc jamais suprieure  10. La boucle est infinie.

````py
# Programme 4
i = 0
while i<1000000 :
    i = i+0.001
````

Le programme 4 se termine pas car la valeur de la variable `i` est augmente de 0.001  chaque passage dans la boucle et finira donc par dpasser 1 000 000.

`````python
# Programme 4
i = 0
while i<1000000 :
    i = i+0.001
`````

## Exercice 5

crire un programme qui simule le lancer de trois ds et qui recommence jusqu ce quon obtienne un triple 6, et qui affiche le nombre dessais qui ont t ncessaires pour obtenir ce rsultat, de la manire prsente ci-dessous :

````pycon
>>> 
Jobtiens 5 4 2 Perdu !
Jobtiens 6 1 6 Perdu !
. . .
Jobtiens 6 6 6 Gagn en 124 essais !
````

**Indication**

La fonction `randint` de la bibliothque `random` gnre un entier alatoire compris au sens large entre les deux entiers donns en argument. Par exemple `random(8,18)` renvoie un entier alatoire compris au sens large entre 8 et 18. Pour pouvoir tre utilise, cette fonction doit tre importe en tapant par exemple `from random import randint` en dbut de programme.

## Corrig de l'exercice 5

`````python
from random import randint

compteur = 0
perdu = True
while perdu:
    de1, de2, de3 = randint(1,6), randint(1,6), randint(1,6)
    compteur = compteur + 1
    if de1 == 6 and de2 == 6 and de3 == 6:
        perdu = False
        phrase = f"J'obtiens {de1} {de2} {de3} Gagn en {compteur} essais !"
    else:
        phrase = f"Jobtiens {de1} {de2} {de3} Perdu !"
    print(phrase)
`````

## Exercice 6

Le client dune banque se prsente devant un automate pour consulter son relev de compte.

Pour y avoir accs, il doit saisir son code personnel qui est compos de 4 chiffres. Il devra ensuite confirmer sa saisie.

crire un programme qui demande  lutilisateur de saisir son code, puis de confirmer sa saisie. Si les deux saisies sont diffrentes, un message derreur sera affich et il devra recommencer, et si au bout de 10 essais il na toujours pas russi, on lui dira que laccs est refus.

## Corrig de l'exercice 6

`````python
erreur = True
compteur = 0
while erreur and compteur < 10:
    compteur = compteur + 1
    print(f"Essai n {compteur} sur 10")
    saisie1 = input("Veuillez saisir votre code : ")
    saisie2 = input("Veuillez confirmer votre saisie : ")
    if saisie1 == saisie2:
        print("Saisie russie !")
        erreur = False
    else:
        "Erreur de code ! Veuillez ressayer !"
if erreur:
    print("Accs refus !")
`````

_Dans cet exercice, rpondre  chaque question en crivant une programme utilisant une boucle `for`, puis un programme utilisant une boucle `while`._

2. crire un programme qui affiche une suite de 12 nombres dont chacun est gal au triple du
prcdent.

## Corrig exercice 10

`````python
# Question 1
# Version avec for

for k in range(20):
    print(k*6)

# Version avec while

k = 0
while k < 20:
    print(k*6)
    k = k + 1
`````

`````python
# Question 2

# Version avec for

nombre = 1
for k in range(12):
    print(nombre)
    nombre = nombre * 3

# Version avec while

nombre = 1
k = 0
while k < 12:
    print(nombre)
    nombre = nombre * 3
    k = k + 1
`````
